```python
X = dataset2.iloc[:,0:13]
```

```python
X #input
```

```python
y = dataset2[["target"]]
y.head()
```

```python
#Train test split

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.80, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = RandomForestClassifier(n_estimators = 200) #means it will implemetn total 150 decision tree behind the scene
```

```python
model1.fit(X_train,y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```
